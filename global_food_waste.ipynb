```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/global_food_wastage_dataset.csv")
df.head() #display first 5 records
```

```python
df.drop(columns=["Food Category"],inplace=True)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
for i in ["Country"]:
  df[i] = label_encoder.fit_transform(df[i])
```

```python
df.head()
```

```python
x=df.iloc[:,0:8]
x
```

```python
df.sort_values(by=['Country'], ascending=True, inplace=True)
```

```python
df.head()
```

```python
y=df['Household Waste (%)']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
ypred=model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,r2_score
mae=mean_absolute_error(y_test,ypred)
r2=r2_score(y_test,ypred)
```

```python
print("Mean Absolute Error:",mae)
print("R-squared Score:",r2)
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor()
model.fit(x_train,y_train)

ypred=model.predict(x_test)
from sklearn.metrics import mean_absolute_error,r2_score
mae=mean_absolute_error(y_test,ypred)
r2=r2_score(y_test,ypred)
print("Mean Absolute Error:",mae)
print("R-squared Score:",r2)
```

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()
model.fit(x_train,y_train)

ypred=model.predict(x_test)
from sklearn.metrics import mean_absolute_error,r2_score
mae=mean_absolute_error(y_test,ypred)
r2=r2_score(y_test,ypred)
print("Mean Absolute Error:",mae)
print("R-squared Score:",r2)
```
